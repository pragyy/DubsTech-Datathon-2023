```python
# importing data manipulation packages
import pandas as pd
import numpy as np

# importing packages for data visualization
import seaborn as sns
import matplotlib.pyplot as plt
```

## Cleaning Data 1

```python
# importing the dataset
df = pd.read_csv('.\Data\sales_data_2017_2018.csv')
df.head()
```

```python
# describing the data 
df.describe()
```

```python
# data infomration
df.info()
```

```python
# performing null check
df.isnull().sum()
```

```python
# looking for duplicates
df.duplicated().any()
```

```python
#identify duplicate rows
duplicateRows = df[df.duplicated()]
duplicateRows
```

upon investigating duplicated rows we can see that the `receipt_id` and `date` attributes are different.

```python
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
df['unit_buying_price'].unique()
```

```python
# filling in missing values
df['unit_buying_price'] = df['unit_selling_price'] - df['unit_price_margin']
df['total_buying_price'] = df['total_selling_price'] - df['total_profit']
```

```python
# performing null check
df.isnull().sum()
```

```python
df.to_csv('./Data/Cleaned Data/data1.csv')
```
